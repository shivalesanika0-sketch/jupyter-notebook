```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'math': [80, 90, 70],
    'science': [85, 95, 75],
    'english': [78, 88, 98]
}

df = pd.DataFrame(data)

corr_matrix = df.corr()

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'math': [80, 90, 70],
    'science': [85, 95, 75],
    'english': [78, 88, 98]
}

df = pd.DataFrame(data)

plt.figure(figsize=(8,5))
for subject in df.columns:
    plt.plot(df.index + 1, df[subject], marker='o', label=subject)

plt.title("Student Marks Line Graph")
plt.xlabel("Student Number")
plt.ylabel("Marks")
plt.xticks(df.index + 1)  
plt.ylim(0, 100)          
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = {
    'math': [80, 90, 70],
    'science': [85, 95, 75],
    'english': [78, 88, 98]
}

df = pd.DataFrame(data)
students = [f"Student {i+1}" for i in df.index]

bar_width = 0.25
r1 = np.arange(len(students))
r2 = [x + bar_width for x in r1]
r3 = [x + bar_width for x in r2]

plt.figure(figsize=(8,5))
plt.bar(r1, df['math'], width=bar_width, label='Math')
plt.bar(r2, df['science'], width=bar_width, label='Science')
plt.bar(r3, df['english'], width=bar_width, label='English')

plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Students' Marks Bar Chart")
plt.xticks([r + bar_width for r in range(len(students))], students)
plt.ylim(0, 100)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

marks = [80, 90, 70, 85, 95, 75, 78, 88, 98]

plt.hist(marks, bins=5, color='skyblue', edgecolor='black')
plt.title("Histogram of Students' Marks")
plt.xlabel("Marks")
plt.ylabel("Number of Students")
plt.show()
```
